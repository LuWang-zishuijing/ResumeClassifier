# 2-1-2. try k-means clustering - max_features=1000  based on cleaning csv data, limited dataset 

```python
actors_name = pd.read_csv('./data/actors_70s.csv')
actors_name.head()
```

```python
# Limit the max_features 
vectorizer = TfidfVectorizer(max_features=1000)
X_limited = vectorizer.fit_transform(test_create_csv_df['words'])
features_limited = vectorizer.get_feature_names()
```

```python
kmeans_max_features_c8 = KMeans()
kmeans_max_features_c8.fit(X_limited)
top_centroids = kmeans_max_features_c8.cluster_centers_.argsort()[:,-1:-11:-1]
for num, centroid in enumerate(top_centroids):
    print("%d: %s" % (num, ", ".join(features_limited[i] for i in centroid)))
```

```python
print("random sample of name_id in each cluster")
assigned_cluster = kmeans_max_features_c8.transform(X_limited).argmin(axis=1)
# print(kmeans_max_features.n_clusters)

for i in range(kmeans_max_features_c8.n_clusters):
    cluster = np.arange(0, X_limited.shape[0])[assigned_cluster==i]
    sample_name_id = np.random.choice(cluster, 5, replace=False)
    print("cluster %d:" % i)
    for name_id in sample_name_id:
        actor_name_id = test_create_csv_df.loc[name_id]['actor_id']
        name = actors_name[ actors_name['name_id'] == actor_name_id ]['name'].values

        print( "       {}, {}. ".format(actor_name_id, name) )
```

```python
kmeans_max_features_c10 = KMeans(n_clusters=10)
kmeans_max_features_c10.fit(X_limited)
top_centroids = kmeans_max_features_c10.cluster_centers_.argsort()[:,-1:-11:-1]
for num, centroid in enumerate(top_centroids):
    print("%d: %s" % (num, ", ".join(features_limited[i] for i in centroid)))
```

```python
print("random sample of name_id in each cluster")
assigned_cluster = kmeans_max_features_c10.transform(X_limited).argmin(axis=1)
# print(kmeans_max_features.n_clusters)

for i in range(kmeans_max_features_c10.n_clusters):
    cluster = np.arange(0, X_limited.shape[0])[assigned_cluster==i]
    sample_name_id = np.random.choice(cluster, 3, replace=False)
    print("cluster %d:" % i)
    for name_id in sample_name_id:
        actor_name_id = test_create_csv_df.loc[name_id]['actor_id']
        name = actors_name[ actors_name['name_id'] == actor_name_id ]['name'].values

        print( "       {}, {}. ".format(actor_name_id, name) )
```

```python
tfidf_weight = X_limited.toarray()
tfidf_weight
```

```python
# Dimensionality Reduction : T-SNE algorithm 
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

tsne = TSNE(n_components=2)
decomposition_data = tsne.fit_transform(tfidf_weight)

fig = plt.figure(figsize=(10, 10))

for color in range(0,10):
    label = []
    x = []
    y = []
    
    for i in range(len(kmeans_max_features_c10.labels_)):
        if kmeans_max_features_c10.labels_[i] == color:
            x.append(decomposition_data[i][0])
            y.append(decomposition_data[i][1])

    plt.scatter(x, y, label=color, marker="x")

# plt.xticks(())
# plt.yticks(())
plt.legend()
plt.savefig('./images/kmeans_1000features_10clusters.png', aspect=1)

```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

```python
# The Elbow Method
```

```python
def within_cluster_sum_of_squares(X, y, cluster):
    cluster_mean = X[y==cluster].mean(axis=0)
    
    return (np.linalg.norm((X[y==cluster] - cluster_mean), axis=1)**2).sum()
```

```python
def sum_within_cluster_sum_of_squares(X, y):
    tmp = []
    for i in range(max(y)+1):
        tmp.append( within_cluster_sum_of_squares(X,y,i) )
    return sum( tmp )
```

```python
nrows, ncols = 3, 4
# fig, axs = plt.subplots(nrows, ncols, figsize=(12,8))

wcss = []
silhouette = []
for i in range(nrows):
    for j in range(ncols):
        k = i*ncols+j + 1
        y = KMeans(k).fit_predict(X_limited)
        
        wcss.append( sum_within_cluster_sum_of_squares(X_limited,y) )
        if k>1:
            silhouette.append( silhouette_score(X_limited,y) )
        else:
            silhouette.append(0)
            
print(silhouette)
```

```python
fig, ax = plt.subplots()
ax.plot(range(1,len(wcss)+1), wcss, 'o-')
ax.set_xlabel("number of clusters")
ax.set_ylabel("within-cluster sum of squares")
```

```python
fig, ax = plt.subplots()
ax.plot(range(2,13), silhouette[1:12], 'o-')
ax.set_xlabel("number of clusters")
ax.set_ylabel("silhouette score")
#ax.set_ylim(ymin=0.0, ymax=1.0)
```
