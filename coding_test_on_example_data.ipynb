# 2-1-2. try k-means clustering - max_features=1000  based on cleaning csv data, limited dataset 

```python
# Limit the max_features 
vectorizer = TfidfVectorizer(max_features=1000)
X_limited = vectorizer.fit_transform(test_create_csv_df['words'])
features_limited = vectorizer.get_feature_names()
```

```python
kmeans_max_features = KMeans()
kmeans_max_features.fit(X_limited)
top_centroids = kmeans_max_features.cluster_centers_.argsort()[:,-1:-11:-1]
for num, centroid in enumerate(top_centroids):
    print("%d: %s" % (num, ", ".join(features_limited[i] for i in centroid)))
```

```python
actors_name = pd.read_csv('actors_70s.csv')
actors_name.head()
```

```python
print("\n5) random sample of name_id in each cluster")
assigned_cluster = kmeans_max_features.transform(X_limited).argmin(axis=1)
# print(kmeans_max_features.n_clusters)

for i in range(kmeans_max_features.n_clusters):
    cluster = np.arange(0, X_limited.shape[0])[assigned_cluster==i]
    sample_name_id = np.random.choice(cluster, 10, replace=False)
    print("cluster %d:" % i)
    for name_id in sample_name_id:
        actor_name_id = test_create_csv_df.loc[name_id]['actor_id']
        name = actors_name[ actors_name['id'] == actor_name_id ]['name'].values

        print( "       {}, {}. ".format(actor_name_id, name) )
```
